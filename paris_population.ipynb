```python
import pandas as pd
import pymysql
import re

conn = pymysql.connect(
    host='localhost',
    user='agfirass',
    password='001291100',
    database='open_data_paris'
)

df = pd.read_sql("SELECT * FROM paris_population", conn)

df_long = df.melt(
    id_vars=['arrondissement'],
    value_vars=['pop_2018', 'pop_2019', 'pop_2020', 'pop_2021', 'pop_2022', 'pop_2023'],
    var_name='year',
    value_name='population'
)

df_long['year'] = df_long['year'].str.replace('pop_', '').astype(int)

def arrondissement_to_number(x):
    match = re.search(r'\d+', x)
    return int(match.group()) if match else None

df_long['arrondissement'] = df_long['arrondissement'].apply(arrondissement_to_number)

df_long = df_long.sort_values(['arrondissement', 'year']).reset_index(drop=True)

df_long.head(50)
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    'mysql+pymysql://agfirass:001291100@localhost/open_data_paris'
)

df_long.to_sql(
    name='paris_population',
    con=engine,
    if_exists='replace',
    index=False
)
```
